```python
import joblib
import pandas as pd

from sklearn.metrics import classification_report

data=pd.read_csv('./data/train.csv')
test=pd.read_csv('./data/unseen.csv')
data.shape,test.shape
```

```python
cls=joblib.load('./model/ridge_model.pkl')
cls
```

```python
unseen_X=test.drop('Loan_Status',axis=1)
y=test['Loan_Status']
```

```python
y_pred=cls.predict(unseen_X)
y_pred
```

```python
print(classification_report(y,y_pred))
```
