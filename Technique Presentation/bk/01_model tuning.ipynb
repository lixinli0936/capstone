```python
import pandas as pd

data= pd.read_csv('./data/train.csv')
data.shape
```

```python
data.head()
```

```python
from pycaret.classification import ClassificationExperiment
s=ClassificationExperiment()
```

```python
s.setup(data,target='Loan_Status',session_id=930)
```

```python
best=s.compare_models()
```

```python
s.evaluate_model(best)
```

```python
pred_holdout=s.predict_model(best)
```

```python
new_data=data.copy().drop('Loan_Status',axis=1)
prediction=s.predict_model(best,data=new_data)
```

```python
from sklearn.metrics import classification_report
y=data['Loan_Status']
y_pred=prediction['prediction_label']
print(classification_report(y,y_pred))
```

```python
test=pd.read_csv('./data/unseen.csv')
new_data=test.copy().drop('Loan_Status',axis=1)
prediction=s.predict_model(best,data=new_data)
y=test['Loan_Status']
y_pred=prediction['prediction_label']
print(classification_report(y,y_pred))
```

```python
ridge=s.create_model('ridge')
```

```python
tuned_ridge=s.tune_model(ridge)
tuned_ridge
```

```python
final_ridge=s.finalize_model(tuned_ridge)
final_ridge
```

```python
s.save_model(final_ridge,'./model/ridge_model')
```
